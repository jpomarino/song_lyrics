```python
genius = lyricsgenius.Genius(
    genius_api_key,
    skip_non_songs=True,
    excluded_terms=["(Remix)", "(Live)", "(Demo)", "(Instrumental)"],
    remove_section_headers=True,
    verbose=False  # silences the per-song print statements
)
```

```python
artists = ['Audrey Hobert', 'Holly Humberstone']
```

```python
def get_song_details(genius, song_id):
    
    try:
        response = genius.song(song_id)
        song_data = response["song"]
        tags = [tag["name"] for tag in song_data.get("tags", [])]
        album_info = song_data.get("album")
        album_name = album_info["name"] if album_info else None
        return album_name, tags
    except Exception as e:
        print(f"  Could not fetch details for song ID {song_id}: {e}")
        return None, []
```

```python
import time

records = []

for artist_name in artists:
    print(f"Fetching discography for: {artist_name}")
    
    try:
        artist = genius.search_artist(artist_name, max_songs=None)
    except Exception as e:
        print(f"  Failed to fetch artist {artist_name}: {e}")
        continue

    for song in artist.songs:
        if not song.lyrics:
            continue

        records.append({
            "artist": artist_name,
            "title": song.title,
            "album": song.album['name'],
            "release_date": song.album['release_date_for_display'],
            "lyrics": song.lyrics
        })

        time.sleep(0.5)  # stay polite to the API

    print(f"  Collected {len([r for r in records if r['artist'] == artist_name])} songs")
```
